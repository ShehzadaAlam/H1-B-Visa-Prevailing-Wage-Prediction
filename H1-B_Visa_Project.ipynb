## Missing Values
----

```python
# make a list of the variables that contain missing values
vars_with_na = [var for var in dataset.columns if dataset[var].isnull().sum()>1]

# print the variable name and the percentage of missing values
for var in vars_with_na:
    print(var, np.round(dataset[var].isnull().mean(), 3), ' % missing values')
```

---

```python
def analyse_na_value(df, var):
    df = df.copy()
    
    # creating variable that indicates 1 if the observation was missing or zero otherwise
    df[var] = np.where(df[var].isnull(), 1, 0)
     
    # calculate the mean prevailing_wage where the information is missing or present
    df.groupby(var)['prevailing_wage'].mean().plot.bar()
    plt.title(var)
    plt.show()
    
for var in vars_with_na:
    analyse_na_value(dataset, var)
```

```python
#numerical variables
num_vars = [var for var in dataset.columns if dataset[var].dtypes != 'O']

print('Number of numerical variables: ', len(num_vars))

# visualise the numerical variables
dataset[num_vars].head()
```

```python
#Date variables
variables = dataset.columns
date_vars = [var for var in variables if 'year' in var or 'date' in var]
date_vars
```

### Discrete variables
-----

```python
#  list of discrete variables
discrete_vars = [var for var in num_vars if len(dataset[var].unique())<20 and var not in date_vars]

print('Number of discrete variables: ', len(discrete_vars))
```

```python
# Visualise the discrete variables
dataset[discrete_vars].head()
```

```python
def analyse_discrete(df, var):
    df = df.copy()
    df.groupby(var)['prevailing_wage'].median().plot.bar()
    plt.title(var)
    plt.ylabel('prevailing_wage')
    plt.show()
    
for var in discrete_vars:
    analyse_discrete(dataset, var)
```

#### Continuous variables
----

```python
# list of continuous variables
cont_vars = [var for var in num_vars if var not in discrete_vars+date_vars]

print('Number of continuous variables: ', len(cont_vars))
```

```python
# Visualise the continuous variables
dataset[cont_vars].head()
```

```python
# Analyse the distributions of these variables
def analyse_continous(df, var):
    df = df.copy()
    df[var].hist(bins=20)
    plt.ylabel('Frequency')
    plt.xlabel(var)
    plt.title(var)
    plt.show()
    
for var in cont_vars:
    analyse_continous(dataset, var)
```

## Outlier
-----

```python
# boxplots to visualise outliers in the continuous variables 

def find_outliers(df, var):
    df = df.copy()
    df[var] = np.log(df[var])
    df.boxplot(column=var)
    plt.title(var)
    plt.ylabel(var)
    plt.show()
    
for var in cont_vars:
    find_outliers(dataset, var)
```
